# Most commonly used word with more than 5 letters.

```python
counter = {}
for reason in deta['7. Reasons']:
    reason = str(reason)
    split = reason.split()
    for word in split:
        if len(word) > 5:
            if word not in counter:
                counter[word] = 1
            else:
                counter[word] += 1
print(counter)
```
